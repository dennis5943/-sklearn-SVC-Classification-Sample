```python
text = "雖然平衡車在去年大流行的時候有著許多災難般的起火事件，不過被小米與 Ninebot 接手的平衡車界大廠 Segway，旗下的產品倒沒發生過什麼大事件。直到最近，在資安顧問公司 IOActive 的揭露下，才知道 Segway Ninebot Mini Pro（國際版的小米九號平衡車）其實曾是款資安漏洞百出的產品，將可輕易地被有心人士破解完全控制，改變設定、速度、方向或甚至瞬間讓 Mini Pro 失去動力等。簡單的說，就是可以完全透過藍牙遙控的方式掌控你腳下平衡車的一切動作。"
```

```python
import jieba
import jieba.posseg as pseg

seg_list = jieba.cut(text,cut_all=False, HMM=True) 
print(" ".join(seg_list))
```

```python
import pandas as pd
import jieba.posseg as pseg

?pseg.cut
ddict = [{"詞":word,"詞性":flag} for word, flag in pseg.cut(text)]

availkeys = ('詞','詞性')

df = pd.DataFrame(ddict,columns = availkeys)
df
```

```python
#關鍵詞取得
#基于TF-IDF

import jieba.analyse

keywords = jieba.analyse.extract_tags(text,topK = 5,withWeight = True
                                     # ,allowPOS = ('n','nr','ns','a')
                                     )
for item in keywords:    
    print(item[0],item[1])
```
